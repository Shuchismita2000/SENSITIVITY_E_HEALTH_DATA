H(X)

```python
s=0
e=[]
value1=[]
for c in weights.columns:
    for i in range(0,weights.shape[0]):
        s=s+(-(weights[c][i]*math.log(weights[c][i])))
    e.append(s)
    value1.append(s)
    s=0
H={}
for key in weights.columns:
    for value in value1:
        H[key]=value
        value1.remove(value)
        break
```

```python
attribute_entropy_df=pd.DataFrame.from_dict(H,orient='index').rename(columns={0:'Entropy of attribute'})
attribute_entropy_df
```

H(X/Y)

```python
s1=0
e1=[]
value2=[]
x=weights.columns
for c in range(0,weights.shape[1]-1):
    for i in range(0,weights.shape[0]):
        s1=s1+(-((weights[x[c]][i]/weights[x[c+1]][i])*math.log(weights[x[c]][i]/weights[x[c+1]][i])))
    e1.append(s1)
    value2.append(s1)
    s1=0

H1={}
for key in weights.columns:
    for value in value2:
        H1[key]=value
        value2.remove(value)
        break
```

H(Y/X)

```python
s2=0
e2=[]
value3=[]
x=weights.columns
for c in range(0,weights.shape[1]-1):
    for i in range(0,weights.shape[0]):
        s2=s2+(-((weights[x[c+1]][i]/weights[x[c]][i])*math.log(weights[x[c+1]][i]/weights[x[c]][i])))
    e2.append(s2)
    value3.append(s2)
    s2=0
H2={}
for key in weights.columns:
    for value in value3:
        H2[key]=value
        value3.remove(value)
        break
```

IG(X,Y)=H(X)+H(Y)-H(X/Y)-H(Y/X)

```python
Ig=[]
atr=[]
df=[]
x=weights.columns
for i in range (0,weights.shape[1]-1):
    for j in range(i+1,weights.shape[1]-1):
        ig=e[i]+e[j]-e1[i]-e2[i]
        a=x[i]+","+x[j]
        atr.append(a)
        Ig.append(ig)
IG={}
for key in atr:
    for value in Ig:
        IG[key]=value
        Ig.remove(value)
        break
```

 Ui represents user node i


 Di represents request i made by a user.


 Ri represents dataset i.


 Ri,j is specific data item j in dataset Ri.


The edges represent the following:


 cUi;Dj is the number of times user ui has made request Dj


dDi;Rj;k is an edge from request i to specific data item k in dataset Rj.


 dRi;j;Rm;n is an edge from specific data item j in dataset i to specific data
item n in dataset Rm.

```python
def counting_yes( dataset , column_a , column_b ):
    l=[]
    count=0
    c=dataset.columns[column_a:column_b]
    for i in range(0,c.shape[0]):
        for j in range(0,dataset.shape[0]):
            if(dataset[c[i]][j]==1):
                count=count+1
        l.append(count)
        count=0
    b={}
    for key in c:
        for value in l:
            b[key]=value
            l.remove(value)
            break
    return b
```

```python
def counting_no( dataset , column_a , column_b ):
    a=[]
    count=0
    c=dataset.columns[column_a,column_b]
    for i in range(0,c.shape[0]):
        for j in range(0,dataset.shape[0]):
            if(dataset[c[i]][j]==0):
                count=count+1
        a.append(count)
        count=0

    b={}
    for key in c:
        for value in a:
            b[key]=value
            a.remove(value)
            break
    return b
```

```python
d_sum=0
pa_sum=0
ph_sum=0
r_sum=0
for i in D_C_1:
    d_sum=d_sum+D_C_1[i]
for i in PA_C_1:
    pa_sum=pa_sum+PA_C_1[i]
for i in PH_C_1:
    ph_sum=ph_sum+PH_C_1[i]
for i in R_C_1:
    r_sum=r_sum+R_C_1[i]
```

```python
p_atr_doctor=[]
P_atr_doctor={}
for i in D_C_1:
    p=(D_C_1[i])/d_sum
    p_atr_doctor.append(p)
for key in doctor.columns:
    for value in p_atr_doctor:
        P_atr_doctor[key]=value
        p_atr_doctor.remove(value)
        break
```
